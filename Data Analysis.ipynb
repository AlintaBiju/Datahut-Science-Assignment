```python
df['Discounted_percentage'] = ((df['MRP'] - df['Discounted Price']) / df['MRP']) * 100
```

```python
avg_dis = df.groupby('Brand')['Discounted_percentage'].mean()
```

```python
avg_dis.sort_values(ascending = False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df['MRP'], bins=30, kde=False, color='skyblue')
plt.title('Distribution of Product Prices (MRP)')
plt.xlabel('Price')
plt.ylabel('Number of Products')
plt.savefig('product_MRP_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.histplot(df['Discounted Price'], bins=30, kde=False, color='royalblue')
plt.title('Distribution of Product Prices (Discounted price)')
plt.xlabel('Discounted Price')
plt.ylabel('Number of Products')
plt.savefig('product_disountprice_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x= 'Brand',y = 'Avg_Discount_By_Brand',data=df,color='indianred',edgecolor='black')
plt.xlabel("Brand")
plt.ylabel("Avg Discount by Brand")
plt.title("Average discount percentage by brand")
plt.xticks(rotation = 90,ha = 'right')
plt.savefig("avg_discount_percentage_by_brand.png",dpi=300,bbox_inches='tight')
plt.show()
```

```python
df['Avg_Discount_By_Brand'] = df['Brand'].map(avg_dis)
```

```python
sns.boxplot(x='Category', y='MRP', data=df, color='lightseagreen')
plt.title('Price Distribution for category')
plt.ylabel('MRP')
plt.savefig("MRPprice_distribution_for_Mens_casual_shirts.png",dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Discounted_percentage'], df['Rating'], alpha=0.5, color='blue')
plt.title('Ratings vs Discount Percentage')
plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.grid(True,alpha = 0.2)
plt.savefig("rating_vs_discountpercentage.png",dpi=300, bbox_inches='tight')
plt.show()
```
